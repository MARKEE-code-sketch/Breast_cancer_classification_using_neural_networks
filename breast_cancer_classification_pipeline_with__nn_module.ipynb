```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
#dropping column id and unnamed:32
df = df.drop(columns=['id', 'Unnamed: 32'], errors='ignore')
print(df.shape)
```

Test Train split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

Now, scaling the data

```python
scalar=StandardScaler()
X_train=scalar.fit_transform(X_train)
X_test=scalar.transform(X_test)
```

```python
X_test
```

Label Encoding

```python
y_train
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

```python
y_train
```

```python
y_test
```

Transform numpy to tensor

```python
X_train_tensor=torch.from_numpy(X_train).float()
X_test_tensor=torch.from_numpy(X_test).float()
y_train_tensor=torch.from_numpy(y_train).float()
y_test_tensor=torch.from_numpy(y_test).float()
```

```python
X_train_tensor.shape
```

```python
y_train_tensor.shape
```

```python
#using the dataset and dataloader class for training in batches
from torch.utils.data import Dataset,DataLoader
class CustomDataset(Dataset):
 def __init__(self,features,labels):
  self.features=features
  self.labels=labels
 def __len__(self):
  return len(self.features)
 def __getitem__(self,idx):
  return self.features[idx],self.labels[idx]
```

```python
#Creating two objects for training and testing dataset
train_dataset=CustomDataset(X_train_tensor,y_train_tensor)
test_dataset=CustomDataset(X_test_tensor,y_test_tensor)
```

```python
#creating batches for both training and testing dataset
train_loader=DataLoader(train_dataset,batch_size=32,shuffle=True)
test_loader=DataLoader(test_dataset,batch_size=32,shuffle=True)
```

Defining the model

```python
class MySimpleNN(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.linear=nn.Linear(num_features,1)
    self.sigmoid=nn.Sigmoid()

  def forward(self,features):
     out=self.linear(features)
     out=self.sigmoid(out)
     return out
```

Important Parameter


```python
learning_rate=0.1
epochs=25
```

Training pipeline

```python
#create model
model=MySimpleNN(X_train_tensor.shape[1]) #object of class we created

#define optimizer
optimizer=torch.optim.SGD(model.parameters(),lr=learning_rate)

#define loss_function
loss_function=nn.BCELoss()


#training the model

#define loop
for epoch in range(epochs):
  for batch_features,batch_labels in train_loader:
#forward pass
   y_pred = model(batch_features)
#loss calculate
   loss=loss_function(y_pred,batch_labels.view(-1,1))


#zero grad
   optimizer.zero_grad()

#backward pass
   loss.backward()

#parameters update
   optimizer.step()


  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

Model Evaluation

```python
model.eval()  #sets the model in evaluation mode
accuracy_list=[]
#model evaluation
with torch.no_grad():   #evaluation mai backpropagation thodi kr rhe so off kro gradient
 for batch_features,batch_labels in test_loader:
  y_pred = model.forward(batch_features)
 # but hmare y_test_tensor is in form of label 0 & 1, so need to convert it into 0 and 1
  y_pred = (y_pred > 0.5).float()
  print(y_pred)
  #calulate batch accuracy
  batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
  accuracy_list.append(batch_accuracy)
  print(batch_accuracy)


```

```python
sum_accuracy=sum(accuracy_list)
len_accuracy=len(accuracy_list);
overall_accuracy=sum_accuracy/len_accuracy
print(overall_accuracy)
```
